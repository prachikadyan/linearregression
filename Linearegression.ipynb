```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the Iris dataset
iris = datasets.load_iris()

# Select petal length as feature
X = iris.data[:, 2].reshape(-1,1)

# Target variable: petal width
Y = iris.data[:, 3]

print("Feature shape:", X.shape)
print("Target shape:", Y.shape)
```

```python
# Visualize the relationship between petal length and petal width
plt.figure(figsize=(6,4))

plt.scatter(X, Y, color="pink")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Relationship Between Petal Length and Width")

plt.grid()
plt.show()
```

```python
# Split dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
# Create and train the model
model = LinearRegression()

model.fit(X_train, Y_train)

print("Model training completed")
```

```python
# Predict on test data
Y_pred = model.predict(X_test)
```

```python
# Evaluate model performance
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print("Mean Squared Error:", mse)
print("R² Score:", r2)
```

```python
# Plot regression results
plt.figure(figsize=(6,4))

plt.scatter(X, Y, color="pink", label="Actual Data")

# Regression line
plt.plot(X, model.predict(X), color="green", label="Regression Line")

plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Linear Regression on Iris Dataset")

plt.legend()
plt.grid()

plt.show()
```
